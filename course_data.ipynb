```python
start_date = '1985-01-01'
end_date = '2010-01-01'
mask = (df.index >= start_date) & (df.index <= end_date)
data = df.loc[mask]
```

```python
data.plot()
plt.show()
```

```python
yearlydata = data.rolling(window = 12).mean()
yearlydata.plot()
plt.show()
```
